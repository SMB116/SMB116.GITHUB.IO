## Univariate Exploration 

```python
#Change the headers to lower case.
df.columns = df.columns.str.lower()

#What are the type of loan terms that we have? 

sns.countplot(data=df, x='term', color = 'red')
plt.title('Count of Borrowers by Term'), plt.ylabel('Count of Borrowers')
plt.show()
```

As shown above, majority of the borrowers are on a 36 month term.

```python
#what are the loan status of the borrowers? 
sns.countplot(data=df, y='loanstatus', color = 'red') 
plt.title('Borrowers Loan Status'), plt.xlabel('Count of Borrowers'), plt.ylabel('Loan status')
plt.show()
```

As showed above, matority of the loan status are current. However, a little under 40000 are complete. 

```python
#What is the borrowers' annual percentage rate?

plt.hist(data=df, x='borrowerapr', color = 'red');
plt.title('Borrowers Annual Percentage Rate')
plt.xlabel('Borrowers Annual Percentage Rate (in decimals)'), plt.ylabel('Count of Borrower')
plt.show()
```

As showed above, the borrower annual percentage rate ranges from a 5% to a little over 40%. The largerest grouping is around 20% therefore the data is showing that borrower credit scores seems to be fairly good. Further investigation is needed.

```python
#What is the credit score of the borrowers?

#Clean data row titles
df['creditscorerangelower'] = df['creditscorerangelower'].replace(0.0, 300.0)

sns.countplot(data=df, y= 'creditscorerangelower', color = 'red');
plt.title('Lower Value Range Credit scores') 
plt.ylabel('Credit Score Range Lower')
plt.xlabel('Count of Borrowers')
plt.show()
```

Our hypothesis was corrected, a good credit score is considered 670 and above. Our borrowers lower rated credits scores have range from 640 and higher which is consider fairly to good.  

Some changes made to data: Note that credit score value that were zero are replace to have a score of 300 because you cannot have a credit score of zero, 300 is the lowest credit score you can have.

```python
# Are the borrowers homeowners? 
sns.countplot(data=df, x='isborrowerhomeowner', color = 'red');
plt.title('Are the borrowers homeowners?')
plt.xlabel('Homeowners?'), plt.ylabel('Count of Borrower')
plt.show()
```

The results show that there's a little more homeowners in our dataset. Thus, there are more secured loans with the lender since homeowners' properties are used as security for the loans.  

```python
#Clean data row titles
df['incomerange'] = df['incomerange'].replace('Not displayed', 'Prefer Not to Say')
df['incomerange'] = df['incomerange'].replace('$0', 'Not employed')

#What is the range of income for our borrowers?

sns.countplot(data=df, y='incomerange', color = 'red')
plt.title('Borrowers Income Range?')
plt.xlabel('Count of Borrowers'), plt.ylabel('Income Range')
plt.show()
```

The data shows that the majority of the borrowers are below or within and slightly above the average salary in the USA.

Some changes made to data: The income range that was displayed as 'not displayed' rows, we renamed to "Prefer Not to Say'.
Also, we had some borrowers with a income of $0 therefore we group the information to be combined with 'not employed'

```python
#What are the borrowers rate?

plt.hist(data=df, x='borrowerrate', color = 'red');
plt.title('Borrowers Rate')
plt.xlabel('Borrowers Rate (in decimals)'), plt.ylabel('Count of Borrower')
plt.show()
```

Above shows the borrowers rates, further analysis is needed.

## Bivariate Exploration
